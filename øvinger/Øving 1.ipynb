```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
```

```python
def attractive_potential(r, a=0.38,  E=3.2e-20):
    return E*((a/r)**12)

def repulsive_potential(r, a=0.38, E=3.2e-20):
     return E*((a/r)**6)

def johnson_equation(r, a=0.38, E=3.2e-20):
    return attractive_potential(r,a,E) - repulsive_potential(r,a,E)
```
