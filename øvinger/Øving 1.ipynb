```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
```

```python
def attractive_potential(r, a=0.38,  E=3.2e-20):
    return E*((a/r)**12)

def repulsive_potential(r, a=0.38, E=3.2e-20):
     return E*((a/r)**6)

def johnson_equation(r, a=0.38, E=3.2e-20):
    return attractive_potential(r,a,E) - repulsive_potential(r,a,E)
```

## 1 a)

```python
E = 3.2e-20
a = 0.38
r = np.linspace(0.3,1, 5000)

att = E*((a/r)**12)
rep = -E*((a/r)**6)

plt.plot(r, att, r, rep, r, att+rep)
#plt.ylim([-1e-20, 1e-20])
plt.ylim([-1e-20, 1.5e-20])
```
